```python
# 4) check time stamp datatype and extract day in new col
def get_day(s):
    return s.split()[0].split("-")[2]

type(df['timeStamp'].iloc[0]) # str
df['day'] = df['timeStamp'].apply(get_day)
df
```

```python
# 5) filter out all the zip codes whose address starts with 'A'
def starts_with_a(x):
    x = str(x) # not sure why typecast is needed for this column
    return x.startswith('A')

df[df['addr'].apply(starts_with_a)].loc[:,['zip']]
```

```python
# 6) try to map timestamp with day like monday, tuesday....
import datetime as dt

def time_to_day(s):
    s = s.split()[0] # Extract the date
    return dt.datetime.strptime(s, "%Y-%m-%d").strftime("%A") # Convert date to day

df['timeStamp'].apply(time_to_day)
```

```python
# Which transactions have a UserID not in users?
user_txn = users_df.merge(txns_df, on="UserID",how="right")
txn_id_no_users = user_txn[pd.isna(user_txn.User)].loc[:,'TransactionID']
txns_df[txns_df['TransactionID'].isin(txn_id_no_users)]
```
