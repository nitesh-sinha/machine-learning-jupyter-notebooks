```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
# For the same timestamp(example 18) we have 56 datapoints
fmri[fmri['timepoint']==18]
```

```python
# If we plot this dataset with multiple y-axis datapoints for same point on x-axis,
# seaborn will, by default, aggregate the data and plot a line with mean(dark blue line) and 95% confidence
# interval around the mean(shown by blue shadow region)
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line')
```

```python
# This is same plot as above but with confidence interval of 50% around the mean
# ci = 'sd' will mean ci of 1 standard deviation around the mean
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci=50)
```

```python
# Turn off aggregation altogether use estimator=None
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', estimator=None)
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', hue='region', style='event', ci=50)
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', hue='region', style='event', markers=True, dashes=True)
```

```python
# err_style denotes confidence interval drawn as bars or band(default)
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', hue='region', style='event', markers=True, ci=60, err_style='bars')
```

#### 2.2 - Visualizing Distribution of the Data(using seaborn)

```python
# Draws a normalized histogram of probability density function of input data
# rug=False denotes no rug plot(A rug plot is a dataplot for single quantitaive variable.
# Think of a rug plot as a 1-D SCATTER PLOT)
sns.distplot(randn(1000), kde=True, hist=True, rug=False, bins=50)
```

```python
# joint plot(similar to 2-D scatter plot)
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
color_map = sns.cubehelix_palette(as_cmap = True, dark = 0, light = 1, reverse= True)
sns.kdeplot(tips['total_bill'], tips['tip'], cmap=color_map, shade=True, n_levels=50) # n_levels number of contours lines
```

```python
# combine two types of plot
tips_kde_plot = sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde', color='orange')
tips_kde_plot.plot_joint(plt.scatter, marker='+', c='red', s=40)
tips_kde_plot.ax_joint.collections[0].set_alpha(0) ## Not sure what this line does. No visible change in plot without it
```

#### 2.4 - Setting some plot aesthetics

```python
def sinplot():
    x = np.linspace(0,50,100)
    y = np.sin(0.5*x)
    plt.plot(x,y)

sinplot()
```
